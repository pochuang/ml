```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=100)
```

```python
def visualize_cv(cv, X, y):
    fig, ax = plt.subplots(figsize=(10, 5))
    for ii, (tr, tt) in enumerate(cv.split(X, y)):
        p1 = ax.scatter(tr, [ii] * len(tr), c="black", marker="_", lw=8)
        p2 = ax.scatter(tt, [ii] * len(tt), c="red", marker="_", lw=8)
        ax.set(title=cv.__class__.__name__,xlabel="Data Index",ylabel="CV Iteration",ylim=[cv.n_splits, -1],)
    
    ax.legend([p1, p2], ["Training", "Validation"])
    plt.show()
```

```python
from sklearn.model_selection import KFold
cv = KFold(n_splits=5)
visualize_cv(cv, X, y)
```

```python
cv = KFold(n_splits=5, shuffle=True)
visualize_cv(cv, X, y)
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import StratifiedKFold
X, y = make_classification(n_samples=100, n_classes=2)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1121218)
visualize_cv(cv, X, y)
```

```python
from sklearn.model_selection import ShuffleSplit
cv = ShuffleSplit(n_splits=5, train_size=0.75, test_size=0.25)
visualize_cv(cv, X, y)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.5)
visualize_cv(cv, X, y)
```

```python
from sklearn.model_selection import TimeSeriesSplit
cv = TimeSeriesSplit(n_splits=5)
visualize_cv(cv, X, y)
```
