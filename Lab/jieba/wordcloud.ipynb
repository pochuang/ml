```python
import jieba
import jieba.analyse

jieba.set_dictionary(r'C:\pyml_scripts\jieba\dict.txt.big.txt')
jieba.load_userdict(r'C:\pyml_scripts\jieba\kingallnames.txt')

#with open('C:\pyml_scripts\jieba\TL.txt', "rb") as f:
with open(r'C:\pyml_scripts\jieba\TL_cut.txt', "rb") as f:
    for line in f:
        tags = jieba.analyse.extract_tags(line,
                                          topK=10, 
                                          withWeight=True)
        for tag, weight in tags:
            print(tag + "," + str(int(weight)))
```

```python
import codecs
wtags = codecs.open(r'C:\pyml_scripts\jieba\TL_tags.txt', "w","utf-8")

with open(r'C:\pyml_scripts\jieba\TL.txt','r', encoding='utf-8') as f1:
    for line in f1:
        words = jieba.analyse.extract_tags(line,10)
        wtags.write(" ".join(words))
f1.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
stopwords = []
text = open(r'C:\pyml_scripts\jieba\TL_tags.txt','r', encoding='utf-8').read()
wc = WordCloud(background_color="white", #背景顏色
               max_words = 2000 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞
wc.generate(text)
```

```python
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(10,6), dpi = 100)
plt.show()
```
