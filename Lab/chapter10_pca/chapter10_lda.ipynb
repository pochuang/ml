```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris.DESCR)
```

```python
iris.target
```

```python
iris.data
```

```python
plt.scatter(iris.data[0:50,0],iris.data[0:50,1],color='red',marker='*',label='setosa')
plt.scatter(iris.data[50:100,0],iris.data[50:100,1],color='green',marker='+',label='versicolor')
plt.scatter(iris.data[100:,0],iris.data[100:,1],color='blue',marker='x',label='virginica')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca2=PCA(n_components=2)
iris_pca2 = pca2.fit_transform(iris.data)
iris_pca2
```

```python
pca2.explained_variance_ratio_.cumsum()
```

```python
plt.scatter(iris_pca2[0:50,0],iris_pca2[0:50,1],color='red',marker='*',label='setosa')
plt.scatter(iris_pca2[50:100,0],iris_pca2[50:100,1],color='green',marker='+',label='versicolor')
plt.scatter(iris_pca2[100:,0],iris_pca2[100:,1],color='blue',marker='x',label='virginica')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.title('PCA 2')
plt.legend()
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA2 = LinearDiscriminantAnalysis(n_components=2)
```

```python
#ldatarget,.target
iris_lda2=LDA2.fit_transform(iris.data,iris.target)
iris_lda2
```

```python
plt.scatter(iris_lda2[0:50,0],iris_lda2[0:50,1],color='red',marker='*',label='setosa')
plt.scatter(iris_lda2[50:100,0],iris_lda2[50:100,1],color='green',marker='+',label='versicolor')
plt.scatter(iris_lda2[100:,0],iris_lda2[100:,1],color='blue',marker='x',label='virginica')
plt.title('LDA 2')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))
fig1=plt.subplot(121)
plt.scatter(iris_pca2[0:50,0],iris_pca2[0:50,1],color='red',marker='*',label='setosa')
plt.scatter(iris_pca2[50:100,0],iris_pca2[50:100,1],color='green',marker='+',label='versicolor')
plt.scatter(iris_pca2[100:,0],iris_pca2[100:,1],color='blue',marker='x',label='virginica')
plt.title('PCA2')
plt.legend()

fig2=plt.subplot(122)
plt.scatter(iris_lda2[0:50,0],iris_lda2[0:50,1],color='red',marker='*',label='setosa')
plt.scatter(iris_lda2[50:100,0],iris_lda2[50:100,1],color='green',marker='+',label='versicolor')
plt.scatter(iris_lda2[100:,0],iris_lda2[100:,1],color='blue',marker='x',label='virginica')
plt.title('LDA2')
plt.legend()

plt.show()
```
