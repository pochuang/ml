```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
X = np.linspace(0,1,100)
```

```python
def H(x):
    y = []
    for p in x:
        if p == 0:
            h = 0
        else:
            h = -(p*math.log(p,2))
        
        y.append(h)
        
    return y
```

```python
y = H(X)
```

```python
plt.plot(X,y)
plt.xlabel('P(x)')
plt.ylabel('H(x)')
plt.title('H(x)=-P(x)log2(P(x))')
plt.show()
```
