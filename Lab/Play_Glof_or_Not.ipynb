```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import LabelEncoder
```

```python
golfDF=pd.read_csv(r'c:\pyml_scripts\golf.csv')
golfDF
```

```python
le_outlook=LabelEncoder().fit(golfDF['outlook'])
le_outlook.classes_
```

```python
le_outlook_encoder=le_outlook.transform(golfDF['outlook'])
le_outlook_encoder
```

```python
#對golfDF增加一個column:outlook_encoder
golfDF['outlook_encoder']=le_outlook_encoder
golfDF
```

```python
golfDF['temp_encoder']=LabelEncoder().fit_transform(golfDF['temp'])
golfDF['humidity_encoder']=LabelEncoder().fit_transform(golfDF['humidity'])
golfDF['windy_encoder']=LabelEncoder().fit_transform(golfDF['windy'])
golfDF
```

```python
#擷取golfDF數值化的特徵部分
#[:,5:]
#[rows,cols]
X=golfDF.iloc[:,5:]
X
```

```python
y=golfDF['play']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

```python
y_train
```

```python
clfG = GaussianNB() 
clfG.fit(X_train,y_train)
clfG.score(X_test,y_test)
```

```python
predG=clfG.predict(X_test)
predG
```

```python
y_test.values
```

```python
pd.crosstab(y_test.values,predG,rownames=['Actual'],colnames=['Predicated'],margins=True)
```

```python
clfM = MultinomialNB() 
clfM.fit(X_train,y_train)
clfM.score(X_test,y_test)
```

```python
clfB=BernoulliNB()
clfB.fit(X_train,y_train)
clfB.score(X_test,y_test)
```

```python
from sklearn import tree

clfT = tree.DecisionTreeClassifier()
clfT.fit(X_train, y_train)
print("訓練準確度:",clfT.score(X_train, y_train))
print("測試準確度:",clfT.score(X_test, y_test))
```
