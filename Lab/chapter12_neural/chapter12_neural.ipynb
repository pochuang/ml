```python
import numpy as np
import math
import matplotlib.pyplot as plt 
```

```python
def sigmoid(h):
    return 1.0 / (1.0 + np.exp(-h))

def cost(t,o):
    return pow((t-o),2)*0.5    

def reLU(h):
    if h > 0:
        return h
    else:
        return 0
```

```python
Target = 0.0 #Input*Weight+Bias=Target
Input = 1.0
```

```python
Weight = 0.5 #可以使用隨機值代替 
Bias = -1.0  #可以使用隨機值代替
```

```python
Z = Input * Weight + Bias
Output = sigmoid(Z)
Cost = cost(Target,Output)
```

```python
Output
```

```python
Cost
```

```python
X = []
y = []

for i in range(1,1001):
    dCdO = Output-Target
    dOdZ = Output*(1-Output)
    dCdZ = dCdO * dOdZ
    dCdB = dCdZ
    dZdW = 1
    dCdW = dCdZ*dZdW
    LearningRate = 0.5
    Weight = Weight - dCdW * LearningRate
    Bias = Bias - dCdB * LearningRate
    Z = Input * Weight + Bias
    Output = sigmoid(Z)
    Cost = cost(Target,Output)
    print ('%d %10.10f' %(i,Cost))
    X.append(i)
    y.append(Cost)    
```

```python
plt.figure(figsize=(20,10))
plt.plot(X,y)
plt.show()
```

```python
from sklearn import svm
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
wine = pd.read_csv(r'C:\pyml_scripts\chapter12_neural\winequality-red.csv', sep=';')
X = wine[list(wine.columns)[:-1]]
y = wine['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
```

```python
Xs = StandardScaler().fit_transform(X)
Xn = Normalizer().fit_transform(X)
Xm = MinMaxScaler().fit_transform(X)
Xs_train,Xs_test,ys_train,ys_test = train_test_split(Xs,y,test_size=0.1)
Xn_train,Xn_test,yn_train,yn_test = train_test_split(Xn,y,test_size=0.1)
Xm_train,Xm_test,ym_train,ym_test = train_test_split(Xm,y,test_size=0.1)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(10,10,10),max_iter=2000) #3層隱藏層,每個隱藏層有10個神經元,max_iter=500表示訓練500次以期望得到最佳weight與bias
```

```python
mlp.fit(Xs_train,ys_train)
predictions = mlp.predict(Xs_test)
```
