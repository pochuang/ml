```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x1_train = np.array(np.linspace(0,50,300) + 0.5 * np.random.random(300)).reshape(-1,1)
x2_train = np.array(np.linspace(-10,10,300) + 0.02* np.random.random(300)).reshape(-1,1)
```

```python
def y_data(x1,x2):
    y = 0.5 * np.sin(x1) + 0.5 * np.cos(x2) + 2 + 0.1 * x1 + 0.01 * x2
    return y
```

```python
X_train = np.concatenate([x1_train,x2_train],axis=1)
X_train.shape
```

```python
y = []
for x1,x2 in zip(x1_train,x2_train):
    y.append(y_data(x1,x2) + (np.random.random(1) - 0.5))
    
y_train = np.array(y)
y_train.shape
```

```python
x1_test = np.array(np.linspace(0,50,100) + 0.5 * np.random.random(100)).reshape(-1,1)
x2_test = np.array(np.linspace(-10,10,100) + 0.02 * np.random.random(100)).reshape(-1,1)

y = []
for x1,x2 in zip(x1_test,x2_test):
    y.append(y_data(x1,x2) + (np.random.random(1) - 0.5))
    
y_test = np.array(y)

X_test = np.concatenate([x1_test,x2_test],axis=1)
```

```python
x1_new= np.array(np.linspace(50,60,30) + 0.5 * np.random.random(30)).reshape(-1,1)
x2_new = np.array(np.linspace(-10,10,30) + 0.02 * np.random.random(30)).reshape(-1,1)

y_new = []
for x1,x2 in zip(x1_new,x2_new):
    y_new.append(y_data(x1,x2) + (np.random.random(1) - 0.5))
    
y_test_new = np.array(y_new)

X_test_new = np.concatenate([x1_new,x2_new],axis=1)
```

```python
plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,300),y_train,'r.',label='Train Data')
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.legend()
```

```python
from sklearn import linear_model
model_lr = linear_model.LinearRegression()
model_lr.fit(X_train,y_train)
model_lr.score(X_test,y_test)
```

```python
y_pred_lr = model_lr.predict(X_test)

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_lr,'y.-',label='Predict Data')
plt.title('Linear Regression')
plt.legend()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly3 = PolynomialFeatures(3)
X_train_poly3 = poly3.fit_transform(X_train)
X_test_poly3 = poly3.fit_transform(X_test)
from sklearn import linear_model
model_poly3 = linear_model.LinearRegression()
model_poly3.fit(X_train_poly3,y_train)
model_poly3.score(X_test_poly3,y_test)
```

```python
y_pred_poly3 = model_poly3.predict(X_test_poly3)

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_poly3,'y.-',label='Predict Data')
plt.title('Polynominal Regression')
plt.legend()
```

```python
from sklearn.tree import DecisionTreeRegressor
model_dt = DecisionTreeRegressor()
model_dt.fit(X_train,y_train)
model_dt.score(X_test,y_test)
```

```python
y_pred_dt = model_dt.predict(X_test)

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_dt,'y.-',label='Predict Data')
plt.title('Decision Tree')
plt.legend()
```

```python
from sklearn.svm import SVR
model_svr = SVR()
model_svr.fit(X_train,y_train.ravel())
model_svr.score(X_test,y_test)
```

```python
y_pred_svr = model_svr.predict(X_test)

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_svr,'y.-',label='Predict Data')
plt.title('Support Vector Machine Regression')
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn3 = KNeighborsRegressor(n_neighbors=3)
model_knn3.fit(X_train,y_train)
model_knn3.score(X_test,y_test)
```

```python
y_pred_knn3 = model_knn3.predict(X_test)

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_knn3,'y.-',label='Predict Data')
plt.title('KNN 3 neighbors')
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn10 = KNeighborsRegressor(n_neighbors=10)
model_knn10.fit(X_train,y_train)
model_knn10.score(X_test,y_test)
```

```python
y_pred_knn10 = model_knn10.predict(X_test)

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_knn10,'y.-',label='Predict Data')
plt.title('KNN 10 neighbors')
plt.legend()
```

```python
from sklearn.decomposition import PCA
pca1 = PCA(1)
X_train_pca = X_train
X_train_pca1 = pca1.fit_transform(X_train_pca)
X_test_pca = X_test
X_test_pca1 = pca1.fit_transform(X_test_pca)
```

```python
from sklearn.isotonic import IsotonicRegression
model_iso = IsotonicRegression()
model_iso.fit(X_train_pca1.ravel(),y_train.ravel())
y_pred_iso = model_iso.predict(X_test_pca1.ravel())

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_iso,'y.-',label='Predict Data')
plt.title('Isotonic Regression')
plt.legend()
```

```python
from sklearn.ensemble import AdaBoostRegressor
model_ada = AdaBoostRegressor(n_estimators=50)
model_ada.fit(X_train,y_train.ravel())
model_ada.score(X_test,y_test)
```

```python
y_pred_ada = model_ada.predict(X_test)

plt.figure(figsize=(20,15))
plt.plot(np.linspace(0,50,100),y_test,'g.-',label='Test Data')
plt.plot(np.linspace(0,50,100),y_pred_ada,'y.-',label='Predict Data')
plt.title('Ada Boost Regression')
plt.legend()
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rfr = RandomForestRegressor(n_estimators=50)
model_rfr.fit(X_train,y_train.ravel())
model_rfr.score(X_test,y_test)
```
