```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
cluster1=np.random.uniform(0.5,2.5,(2,10))
cluster2=np.random.uniform(2.5,4.5,(2,10))
X=np.hstack((cluster1,cluster2)).T
```

```python
plt.plot(X[:,0],X[:,1],'k.')
plt.show()
```

```python
dbscan = DBSCAN() #min_samples=5,eps=0.5
dbscan_cluster = dbscan.fit_predict(X)
```

```python
dbscan_cluster #-1表示為噪音點
```

```python
colors = ['rd','gx','c+','y*','b^','m>','r.','g+','c*','yx','b<']
```

```python
labels = dbscan_cluster

for i in range(len(X)):
    if labels[i] == -1: 
        plt.plot(X[i][0],X[i][1],'k.')
    else:
        plt.plot(X[i][0],X[i][1],colors[labels[i]])

plt.show()
```

```python
dbscan = DBSCAN(min_samples=3,eps=2) 
dbscan_cluster = dbscan.fit_predict(X)
dbscan_cluster
```

```python
labels = dbscan_cluster

for i in range(len(X)):
    if labels[i] == -1: 
        plt.plot(X[i][0],X[i][1],'k.')
    else:
        plt.plot(X[i][0],X[i][1],colors[labels[i]])

plt.show()
```

```python
dbscan = DBSCAN(min_samples=2,eps=0.5) 
dbscan_cluster = dbscan.fit_predict(X)
dbscan_cluster
```

```python
labels = dbscan_cluster

for i in range(len(X)):
    if labels[i] == -1: 
        plt.plot(X[i][0],X[i][1],'k.')
    else:
        plt.plot(X[i][0],X[i][1],colors[labels[i]])

plt.show()
```

```python
dbscan = DBSCAN(min_samples=2,eps=1) 
dbscan_cluster = dbscan.fit_predict(X)
labels = dbscan_cluster

for i in range(len(X)):
    if labels[i] == -1: 
        plt.plot(X[i][0],X[i][1],'k.')
    else:
        plt.plot(X[i][0],X[i][1],colors[labels[i]])

plt.show()
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_dataset = load_iris()
iris = pd.DataFrame(data= np.c_[iris_dataset['data'], iris_dataset['target']],
                    columns= iris_dataset['feature_names'] + ['target'])
print(iris.columns)
print(list(iris['target']))
```

```python
iris_dbscan_s5 = DBSCAN(eps=0.5,min_samples=5)
iris_dbscan_s5.fit(iris[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)', 'target']])
iris_dbscan_s5_pred = iris_dbscan_s5.labels_
print(iris_dbscan_s5_pred)
```

```python
setosa = iris[iris['target']==0]
versicolor = iris[iris['target']==1]
virginica = iris[iris['target']==2]

#請注意每次所預測的標籤可能與原始標籤順序不相同,所以下面程式碼可能需要變更
setosa_pred = iris[iris_dbscan_s5_pred==0]
versicolor_pred = iris[iris_dbscan_s5_pred==1]
virginica_pred = iris[iris_dbscan_s5_pred==2]
iris_outlier_pred = iris[iris_dbscan_s5_pred==-1]
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.scatter(setosa['petal length (cm)'],setosa['petal width (cm)'],color='red',marker='*',label='setosa')
plt.scatter(versicolor['petal length (cm)'],versicolor['petal width (cm)'],color='green',marker='+',label='versicolor')
plt.scatter(virginica['petal length (cm)'],virginica['petal width (cm)'],color='blue',marker='x',label='virginica')         
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.title('Groud Truth')
plt.legend()

plt.subplot(1,2,2)
plt.scatter(setosa_pred['petal length (cm)'],setosa_pred['petal width (cm)'],color='red',marker='*',label='setosa')
plt.scatter(versicolor_pred['petal length (cm)'],versicolor_pred['petal width (cm)'],color='green',marker='+',label='versicolor')
plt.scatter(virginica_pred['petal length (cm)'],virginica_pred['petal width (cm)'],color='blue',marker='x',label='virginica')
plt.scatter(iris_outlier_pred['petal length (cm)'],iris_outlier_pred['petal width (cm)'],color='black',marker='.',label='Outlier') 
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.title('DBSCAN predict(eps=0.5,min_samples=5)')
plt.legend()

plt.show()
```
