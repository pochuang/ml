```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
data = pd.read_csv(r"C:\pyml_scripts\chapter04_logregression\LRdata2.txt", names=['x1','x2','y'], header=None)
data.head()
```

```python
data0 = data[data['y']==0]
data1 = data[data['y']==1]
plt.figure(figsize=(10,10))
plt.scatter(data0['x1'],data0['x2'],color='r',marker='o')
plt.scatter(data1['x1'],data1['x2'],color='g',marker='+')
plt.show()
```

```python
X = np.array([data['x1'],data['x2']]).T
y = np.array(data['y'])
LR = linear_model.LogisticRegression()
LR.fit(X,y)
LR.score(X,y)
```

```python
LR_intercept = LR.intercept_
[[LR_coef1,LR_coef2]] = LR.coef_
db_x1 = np.array([min(X[:,0]),max(X[:,0])])

def db_x2_fn(coef1,coef2,intercept,x1):
    return ((0-intercept-(coef1*x1))/coef2)

db_x2 = db_x2_fn(LR_coef1,LR_coef2,LR_intercept,db_x1)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data0['x1'],data0['x2'],color='r',marker='o')
plt.scatter(data1['x1'],data1['x2'],color='g',marker='+')
plt.plot(db_x1,db_x2,c='b')
plt.title('C=1.0')
plt.show()
```

```python
from sklearn import preprocessing
poly5 = preprocessing.PolynomialFeatures(5) # 1, x1, x2, x1^2, x1*x2, x2, x1^3, .... x1*x2^4, x2^5 
X = poly5.fit_transform(np.c_[data['x1'],data['x2']]) # np.c_[array1,array2]按照column連接兩個array,np.r_[]則按照row連接
LR_poly5 = linear_model.LogisticRegression(penalty='l2', C=1.0)
LR_poly5.fit(X,y)
```

```python
px = np.arange(-1.0, 1.5, 0.1)
py = np.arange(-1.0, 1.5, 0.1)
PX, PY = np.meshgrid(px, py)
XX = poly5.fit_transform(np.c_[PX.ravel(), PY.ravel()]) 
Z = LR_poly5.predict_proba(XX)[:,1] 
Z = Z.reshape(PX.shape)
plt.figure(figsize=(10,10))
plt.contour(PX, PY, Z, levels=[0.5], linewidths=3,colors='b')
plt.scatter(data0['x1'],data0['x2'],color='r',marker='o')
plt.scatter(data1['x1'],data1['x2'],color='g',marker='+')
plt.title('c=1.0')
plt.show()
```
