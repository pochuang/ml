```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.svm import SVC
```

```python
def plot_decision_regions(x,y,model,resolution=0.02):
    markers = ('s','x','o','^','v')
    colors = ('red','blue','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    x1_min,x1_max = x[:,0].min()-1,x[:,1].max()+1
    x2_min,x2_max = x[:,1].min()-1,x[:,1].max()+1
    xx1,xx2 = np.meshgrid(np.arange(x1_min,x1_max,resolution),
                          np.arange(x2_min,x2_max,resolution))
    z = model.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    z = z.reshape(xx1.shape)
    plt.contourf(xx1,xx2,z,alpha=0.4,cmap=cmap)
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    for idx,c1 in enumerate(np.unique(y)):
        plt.scatter(x=x[y == c1,0],y=x[y == c1,1],
                    alpha=0.8,c=cmap(idx),
                    marker=markers[idx],label=c1)
        
    plt.scatter([2],[-1],c='k',s=100)
    plt.scatter([2],[-3],c='g',s=100)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.show()
```

```python
x_xor = np.random.randn(200,2)
y_xor = np.logical_xor(x_xor[:,0] > 0,x_xor[:,1] > 0)
y_xor = np.where(y_xor,1,0)
```

```python
colors = np.where(y_xor,'blue','red')
```

```python
plt.scatter(x_xor[:,0],x_xor[:,1],color=colors)
plt.show()
```

```python
y_xor
```

```python
svm = SVC(kernel='rbf',C=1.0)
svm.fit(x_xor,y_xor)
plot_decision_regions(x_xor,y_xor,svm)
print(svm.n_support_,svm.score(x_xor,y_xor))
```

```python
svm = SVC(kernel='rbf',gamma=0.1,C=1.0)
svm.fit(x_xor,y_xor)
plot_decision_regions(x_xor,y_xor,svm)
print(svm.n_support_,svm.score(x_xor,y_xor))
```

```python
svm = SVC(kernel='rbf',gamma=1,C=1.0)
svm.fit(x_xor,y_xor)
plot_decision_regions(x_xor,y_xor,svm)
print(svm.n_support_,svm.score(x_xor,y_xor))
```

```python
svm = SVC(kernel='rbf',gamma=10,C=1.0)
svm.fit(x_xor,y_xor)
plot_decision_regions(x_xor,y_xor,svm)
print(svm.n_support_,svm.score(x_xor,y_xor))
```

```python
svm = SVC(kernel='rbf',gamma=100,C=1.0)
svm.fit(x_xor,y_xor)
plot_decision_regions(x_xor,y_xor,svm)
print(svm.n_support_,svm.score(x_xor,y_xor))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_xor,y_xor,test_size=0.2)
svm = SVC(kernel='rbf',gamma=100,C=1.0)
svm.fit(X_train,y_train)
print(svm.n_support_,svm.score(X_test,y_test))
```

```python
svm = SVC(kernel='rbf',gamma=10,C=1.0)
svm.fit(X_train,y_train)
print(svm.n_support_,svm.score(X_test,y_test))
```

```python
svm = SVC(kernel='rbf',gamma=1,C=1.0)
svm.fit(X_train,y_train)
print(svm.n_support_,svm.score(X_test,y_test))
```

```python
svm = SVC(kernel='rbf',C=1.0)
svm.fit(X_train,y_train)
print(svm.n_support_,svm.score(X_test,y_test))
```
