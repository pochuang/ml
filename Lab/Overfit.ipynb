```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(1,11).reshape(10,1)
y = np.array([7,9,7,13,16,14,19,23,18,21]).reshape(10,1)
```

```python
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
print('係數：',model.coef_)
print('截距：',model.intercept_)
```

```python
y_p = model.intercept_ + model.coef_*X
```

```python
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro',X,y_p,'g-')
plt.show()
```

```python
model.score(X,y)
```

```python
X2 = np.c_[X,X**2]
model2 = LinearRegression()
model2.fit(X2,y)
#y_p2 = np.dot(X2,model2.coef_.transpose())+model2.intercept_  #transpose()轉置為1維陣列
y_p2 = model2.predict(X2)
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro',X,y_p2,'b-')
plt.show()
```

```python
model2.score(X2,y)
```

```python
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro',X,y_p,'g-',X,y_p2,'b-')
plt.show()
```

```python
X3 = np.c_[X,X**2,X**3]
model3 = LinearRegression()
model3.fit(X3,y)
y_p3 = np.dot(X3,model3.coef_.transpose())+model3.intercept_
#y_p3 = model3.predict(X3)
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro',X,y_p3,'k-')
plt.show()
```

```python
model3.score(X3,y)
```

```python
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro',X,y_p,'g-',X,y_p2,'b-',X,y_p3,'k-')
plt.show()
```

```python
X9 = np.c_[X,X**2,X**3,X**4,X**5,X**6,X**7,X**8,X**9]
model9 = LinearRegression()
model9.fit(X9,y)
#y_p9 = np.dot(X9,model9.coef_.transpose())+model9.intercept_
y_p9 = model9.predict(X9)
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro',X,y_p9,'y-')
plt.gca()
plt.show()
```

```python
model9.score(X9,y)
```

```python
plt.figure(figsize=(10,10))
plt.plot(X,y,'ro',X,y_p,'g-',X,y_p2,'b-',X,y_p3,'k-',X,y_p9,'y-')
plt.show()
```
