?scatter_matrix用來顯示scatter_matrix的幫助訊息

```python
?scatter_matrix
```

```python
classes = list(iris['target'])
colors_map = {0: 'red', 1: 'green', 2: 'blue'}
colors = [colors_map[c] for c in classes]  
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris.iloc[:,:-1],alpha=0.5,diagonal='hist',figsize=(10,8),color=colors,marker='o')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris.iloc[:,:-1],alpha=0.5,diagonal='kde',figsize=(10,8),color=colors,marker='o')
plt.show()
```

```python
from sklearn.decomposition import PCA
iris_reduced_3 = PCA(n_components=3).fit_transform(iris.iloc[:,:4])
iris_reduced_3[:5]
```

```python
iris.iloc[:5,:4]
```

```python
from mpl_toolkits.mplot3d import Axes3D
#colors = ('r', 'g', 'b')
#c_list = []
#for c in colors:
#    c_list.append([c]*50)

#colors = np.asarray(c_list).flatten()    
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(iris_reduced_3[:,0],iris_reduced_3[:,1],iris_reduced_3[:,2],c=colors)
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.show()
```

```python
from sklearn.decomposition import PCA
iris_reduced_2 = PCA(n_components=2).fit_transform(iris.iloc[:,:4])
iris_reduced_2[:5]
```

```python
plt.scatter(iris_reduced_2[0:50,0],iris_reduced_2[0:50,1],color='red',marker='*',label='山鳶尾花')
plt.scatter(iris_reduced_2[50:100,0],iris_reduced_2[50:100,1],color='green',marker='+',label='變色鳶尾花')
plt.scatter(iris_reduced_2[100:,0],iris_reduced_2[100:,1],color='blue',marker='x',label='維吉尼亞鳶尾花')
plt.legend()
plt.show()
```
