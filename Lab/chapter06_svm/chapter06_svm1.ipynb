```python
import numpy as np  
import matplotlib.pyplot as plt
from sklearn import svm  
```

```python
np.random.seed(1)  
X=np.r_[np.random.randn(20,2)-[2,2],np.random.randn(20,2)+[2,2]]  
Y=[0]*20+[1]*20
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=80,cmap=plt.cm.Paired)
plt.show()
```

```python
clf=svm.SVC(kernel='linear')  #'rbf','poly','sigmoid'
clf.fit(X,Y)
```

```python
clf.coef_
```

```python
w=clf.coef_[0]  
a=-w[0]/w[1]
xx=np.linspace(-5,5)  
yy=a*xx-(clf.intercept_[0])/w[1]  
plt.plot(xx,yy,'k-')
plt.scatter(X[:,0],X[:,1],c=Y,s=80,cmap=plt.cm.Paired)
plt.show()
```

```python
clf.support_vectors_
```

```python
def plot_svc_decision_function(model):
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   c='r',s=100, linewidth=1, facecolors='none');   #繪製support vector
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=80,cmap=plt.cm.Paired)
plot_svc_decision_function(clf)
plt.show()
```

```python
test_x1, test_y1 = (3, 0)
test_x2, test_y2 = (0, 0)
plt.scatter([test_x1],[test_y1],c='g',s=100)
plt.scatter([test_x2],[test_y2],c='y',s=100)
Z1 = clf.predict([[test_x1, test_y1]])
Z2 = clf.predict([[test_x2, test_y2]])
print(Z1)
print(Z2)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=80,cmap=plt.cm.Paired)
plot_svc_decision_function(clf)
plt.scatter([test_x1],[test_y1],c='g',s=100)
plt.scatter([test_x2],[test_y2],c='y',s=100)
plt.show()
```
