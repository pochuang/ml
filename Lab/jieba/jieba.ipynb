```python
import jieba

jieba.set_dictionary(r'C:\pyml_scripts\jieba\dict.txt.big.txt')
jieba.load_userdict(r'C:\pyml_scripts\jieba\kingallnames.txt')

with open(r'C:\pyml_scripts\jieba\TL.txt','r', encoding='utf-8') as f:
    data = [line.strip()
            for line in f.readlines()
            if line.strip()]

sentences = []
for line in data:
    words = list(jieba.cut(line))
    sentences.append(words)
```

```python
sentences[:100]
```

```python
import gensim
model = gensim.models.Word2Vec(sentences,
                               window=5,
                               min_count=5,
                               workers=4)
```

```python
for k, s in model.wv.most_similar(positive=["蕭峰","喬峰"]):
        print('{}:{}'.format(k, s))
```

```python
v = "段譽"
v1 = "段公子"
u = "慕容復"
u1, _ = model.wv.most_similar(positive=[u, v1], negative=[v])[0]
print("如同{0}與{1}的關係,{2}與{3}也有類似關係".format(v, v1, u, u1))
```
