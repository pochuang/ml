```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import scale
from sklearn.preprocessing import Normalizer
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
X = np.array([[2,3],[2,10],[3,5],[4,4],[4,9],[5,7],[6,1],[9,5],[10,8]])
```

```python
X
```

```python
plt.figure(figsize=(10,10))
plt.xlim((-11,11))
plt.ylim((-11,11))
plt.hlines(y=0,xmin=-11,xmax=11)
plt.vlines(x=0,ymin=-11,ymax=11)
plt.scatter(X[:,0],X[:,1])
```

```python
#將數值縮放到 [0-1]之間
minmax = MinMaxScaler()
X_minmax = minmax.fit_transform(X)
X_minmax
```

```python
plt.figure(figsize=(10,10))
plt.title("MinMaxRescale")
plt.xlim((-11,11))
plt.ylim((-11,11))
plt.hlines(y=0,xmin=-11,xmax=11)
plt.vlines(x=0,ymin=-11,ymax=11)
plt.hlines(y=1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=1,ymin=-1,ymax=1,color='yellow')
plt.hlines(y=-1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=-1,ymin=-1,ymax=1,color='yellow')
plt.scatter(X[:,0],X[:,1])
plt.scatter(X_minmax[:,0],X_minmax[:,1])
plt.show()
```

```python
X_sc = scale(X)
X_sc
```

```python
plt.figure(figsize=(10,10))
plt.title("scale")
plt.xlim((-11,11))
plt.ylim((-11,11))
plt.hlines(y=0,xmin=-11,xmax=11)
plt.vlines(x=0,ymin=-11,ymax=11)
plt.hlines(y=1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=1,ymin=-1,ymax=1,color='yellow')
plt.hlines(y=-1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=-1,ymin=-1,ymax=1,color='yellow')
plt.scatter(X[:,0],X[:,1])
plt.scatter(X_sc[:,0],X_sc[:,1])
plt.show()
```

```python
#所有資料的平均值為0，變異數為1
stds = StandardScaler()
X_stds = stds.fit_transform(X)
X_stds
```

```python
plt.figure(figsize=(10,10))
plt.title("Standardization")
plt.xlim((-11,11))
plt.ylim((-11,11))
plt.hlines(y=0,xmin=-11,xmax=11)
plt.vlines(x=0,ymin=-11,ymax=11)
plt.hlines(y=1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=1,ymin=-1,ymax=1,color='yellow')
plt.hlines(y=-1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=-1,ymin=-1,ymax=1,color='yellow')
plt.scatter(X[:,0],X[:,1])
plt.scatter(X_stds[:,0],X_stds[:,1])
plt.show()
```

```python
#將資料縮放到單位範數：1(unit vector的長度)
norms = Normalizer()
X_norms = norms.fit_transform(X)
X_norms
```

```python
plt.figure(figsize=(10,10))
plt.title("Normalization")
plt.xlim((-11,11))
plt.ylim((-11,11))
plt.hlines(y=0,xmin=-11,xmax=11)
plt.vlines(x=0,ymin=-11,ymax=11)
plt.hlines(y=1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=1,ymin=-1,ymax=1,color='yellow')
plt.hlines(y=-1,xmin=-1,xmax=1,color='yellow')
plt.vlines(x=-1,ymin=-1,ymax=1,color='yellow')
plt.scatter(X[:,0],X[:,1])
plt.scatter(X_norms[:,0],X_norms[:,1])
plt.show()
```
